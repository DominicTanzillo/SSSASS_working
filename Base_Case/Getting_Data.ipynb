```python
WinSize = 3200 # .1 second
current_idy = 42000
Y = n1[41500:current_idy+WinSize,3]
Y
```

```python
def update_Y(Y,current_idy,WinSize,n1):
    Y = np.concatenate((Y[-500:],n1[current_idy:current_idy+WinSize,3]))
    current_idy += WinSize
    return Y, current_idy
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
x2 = range(3700)
ax.plot(x2,Y)
current_idy += WinSize
```
